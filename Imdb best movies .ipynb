```python
from bs4 import BeautifulSoup
import requests
movie_page = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
```

```python
soup=BeautifulSoup(movie_page.text,"html.parser")

```

```python
Name=[]
Year=[]
Rating=[]
for i in soup.find_all('div', attrs={'class':'lister-item-content'}):
    Name.append(i.find("h3",attrs={"class":"lister-item-header"}).a)
    Year.append(i.find("span",attrs={"class":"lister-item-year text-muted unbold"}))
    Rating.append(i.find("div",attrs={"class":"inline-block ratings-imdb-rating"}))
```

```python
data=list(zip(Name,Year,Rating))
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=["Name","Year","rating"])
```
